```python
def getX(dataset,tamvarY):
    n= len(dataset[0])
    salida = []
    for a in dataset:
        aux=a[0:n-tamvarY]
        #for i in range(tamvarY): aux.append(1)
        aux.append(1)
        salida.append(aux)
    return salida
def getY(dataset,tamvarY):
    return [np.array(dat[len(dataset[0]) - tamvarY: len(dataset[0])]) for dat in dataset]
```
